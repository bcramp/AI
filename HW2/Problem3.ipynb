**The problem:**

Navigate through mazes with DFS if the mazes provided have a successfull path to traverse.

```python
#@title ExploreMaze Function

# Import the Deque class from the Collections library
from collections import deque

# The ExploreMaze function traverses through a maze provided and prints them out
def ExploreMaze(maze, start, goal):
    # Create a visited set and add the desired start (root node) first
    visited = set()
    visited.add(start)

    # Create an active queue (using deque) to track the active nodes and add the desired start (root node)
    active = deque([start])

    # Keep exploring as long as the queue is not empty
    while active:
        # Dequeue a node from the active queue
        node = active.pop()

        # Iterate through the node's neighbors
        # The reversed() method is utilized to make sure ordering from left-to-right
        for neighbor in reversed(maze[node]):
            # If the node has not been visited, add it to the visited set
            if neighbor not in visited:
                visited.add(neighbor)
                active.append(neighbor)

            # If the goal state has been found, break the loop
            if goal in visited:
                break

    # Have a check that the goal state had FAILED to be found
    if goal not in visited:
        print("-----------------------")
        print("FAILED to find the goal state of {g} starting from {s}!".format(g=goal, s=start))
        print("-----------------------")
    else:
        # Print that the goal state was SUCCESSFULLY found
        print("-----------------------")
        print("SUCCESSFULLY found the goal state of {g} starting from {s}!".format(g=goal, s=start))
        print("-----------------------")
```
